## Part 2: IAM Primitive Roles

```python
# IAM Primitive Roles Overview

iam_roles = {
    "roles/viewer": {
        "description": "Read-only access to all resources",
        "can_view": True,
        "can_edit": False,
        "can_delete": False,
        "can_manage_iam": False
    },
    "roles/editor": {
        "description": "View, create, modify, delete resources",
        "can_view": True,
        "can_edit": True,
        "can_delete": True,
        "can_manage_iam": False
    },
    "roles/owner": {
        "description": "Full control including IAM management",
        "can_view": True,
        "can_edit": True,
        "can_delete": True,
        "can_manage_iam": True
    }
}

print("=== IAM Primitive Roles ===\n")
for role, perms in iam_roles.items():
    print(f"{role}")
    print(f"  Description: {perms['description']}")
    print(f"  View: {'✓' if perms['can_view'] else '✗'}")
    print(f"  Edit: {'✓' if perms['can_edit'] else '✗'}")
    print(f"  Delete: {'✓' if perms['can_delete'] else '✗'}")
    print(f"  Manage IAM: {'✓' if perms['can_manage_iam'] else '✗'}")
    print()
```

## Part 3: Predefined vs Custom Roles

```python
# Role Types Comparison

role_types = {
    "Primitive Roles": {
        "examples": ["roles/owner", "roles/editor", "roles/viewer"],
        "scope": "Project-level only",
        "use_case": "Simple permission management"
    },
    "Predefined Roles": {
        "examples": ["roles/bigquery.dataViewer", "roles/storage.objectAdmin", "roles/aiplatform.user"],
        "scope": "Resource-level granularity",
        "use_case": "Fine-grained access control"
    },
    "Custom Roles": {
        "examples": ["Custom: Project Viewer + BigQuery"],
        "scope": "Organization or Project level",
        "use_case": "Tailored permissions for specific needs"
    }
}

print("=== Role Types Comparison ===\n")
for role_type, details in role_types.items():
    print(f"{role_type}:")
    print(f"  Examples: {', '.join(details['examples'][:2])}")
    print(f"  Scope: {details['scope']}")
    print(f"  Best for: {details['use_case']}")
    print()
```

## Part 4: BigQuery - ARRAY_AGG & STRUCT (SQL Track)

```python
# BigQuery ARRAY_AGG and STRUCT examples

# ARRAY_AGG: Aggregates values into an array
array_agg_example = """
-- Get all products purchased by each customer
SELECT 
    customer_id,
    ARRAY_AGG(product_name) as products_bought
FROM orders
GROUP BY customer_id
"""

# STRUCT: Creates a structured (nested) data type
struct_example = """
-- Create a STRUCT with multiple fields
SELECT 
    STRUCT(
        first_name, 
        last_name, 
        age
    ) as person_info
FROM users
"""

# Combined: ARRAY_AGG with STRUCT
combined_example = """
-- Get order details as array of structs
SELECT 
    customer_id,
    ARRAY_AGG(
        STRUCT(
            order_id,
            order_date,
            total_amount
        )
    ) as orders
FROM orders
GROUP BY customer_id
"""

print("=== BigQuery ARRAY_AGG & STRUCT ===\n")
print("1. ARRAY_AGG Example:")
print(array_agg_example)
print("\n2. STRUCT Example:")
print(struct_example)
print("\n3. Combined (ARRAY_AGG + STRUCT):")
print(combined_example)
```
