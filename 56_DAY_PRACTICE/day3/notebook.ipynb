## 2️⃣ Default Parameters

````python
# Default parameter - use this when you have a sensible default
def greet_person(name, greeting="Hello"):
    return f"{greeting}, {name}!"

# Using default
print(greet_person("Krishna"))

# Overriding default
print(greet_person("Krishna", greeting="Namaste"))
````

````python
def greet(name,greeting="Hello"):
    return f"{greeting},{name}!"
print(greet("krishna")) 
print(greet("krishna",greeting="Namaste"))
````

## 3️⃣ *args — Multiple Arguments

````python
# *args accepts unlimited positional arguments
# args becomes a tuple
def sum_all(*args):
    print(f"Received: {args}")
    total = 0
    for num in args:
        total = total + num
    return total

print("Sum of 1,2,3:", sum_all(1, 2, 3))
print("Sum of 10,20,30,40:", sum_all(10, 20, 30, 40))
````

## 4️⃣ **kwargs — Keyword Arguments

````python
# **kwargs accepts unlimited keyword arguments
# kwargs becomes a dictionary
def configure_model(**kwargs):
    print("Configuration:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

configure_model(
    model='gemini-2.0-flash',
    temperature=0.7,
    max_tokens=1000
)
````

### 🎯 Today's SQL Topic: CTEs (Common Table Expressions)

A CTE is like a temporary table that exists only for one query. Use `WITH` keyword.

---

### SQL 1: CTE with StackOverflow Data

```python
# Query 1: Monthly question count using CTE
query1 = """
WITH monthly_questions AS (
  SELECT 
    DATE_TRUNC(creation_date, MONTH) as month,
    COUNT(*) as question_count
  FROM `bigquery-public-data.stackoverflow.posts_questions`
  WHERE creation_date BETWEEN '2024-01-01' AND '2024-12-31'
  GROUP BY 1
)
SELECT 
  month,
  question_count
FROM monthly_questions
WHERE question_count > 50000
ORDER BY month
"""

# Run query
df1 = client.query(query1).to_dataframe()
print("📊 Monthly Questions (CTE):")
print(df1.head(10))
```

### SQL 2: Nested CTE - Multiple Transformations

```python
# Query 2: CTE inside CTE - Find top tags per month
query2 = """
WITH 
questions AS (
  SELECT 
    id,
    title,
    tags,
    creation_date
  FROM `bigquery-public-data.stackoverflow.posts_questions`
  WHERE creation_date >= '2024-01-01'
  LIMIT 10000
),
monthly_tag_stats AS (
  SELECT 
    DATE_TRUNC(creation_date, MONTH) as month,
    SPLIT(tags, '|') as tag_array,
    COUNT(*) as question_count
  FROM questions
  GROUP BY 1
)
SELECT month, question_count
FROM monthly_tag_stats
ORDER BY month
LIMIT 10
"""

df2 = client.query(query2).to_dataframe()
print("📊 Tag Stats (Nested CTE):")
print(df2)
```

### SQL 3: CTE with Window Function

```python
# Query 3: CTE + Window Function - Rank months by activity
query3 = """
WITH monthly_activity AS (
  SELECT 
    DATE_TRUNC(creation_date, MONTH) as month,
    COUNT(*) as total_questions,
    SUM(answer_count) as total_answers
  FROM `bigquery-public-data.stackoverflow.posts_questions`
  WHERE creation_date >= '2023-01-01'
  GROUP BY 1
),
ranked_months AS (
  SELECT 
    month,
    total_questions,
    total_answers,
    RANK() OVER (ORDER BY total_questions DESC) as rank
  FROM monthly_activity
)
SELECT 
  month,
  total_questions,
  total_answers,
  rank
FROM ranked_months
WHERE rank <= 10
ORDER BY rank
"""

df3 = client.query(query3).to_dataframe()
print("📊 Top 10 Months by Activity (CTE + Window):")
print(df3)
```

### 📚 Key CTE Patterns

| Pattern | Use Case |
|---------|----------|
| `WITH x AS (...)` | Simple CTE |
| `WITH x AS (), y AS ()` | Chained CTEs |
| `WITH cte AS (...) SELECT * FROM cte` | CTE + main query |
| CTE + Window Function | Ranking, running totals |
