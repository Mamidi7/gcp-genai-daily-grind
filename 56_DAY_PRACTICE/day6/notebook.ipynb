## 5. Demo: python-dotenv

```python
# Create a .env file
env_content = """GEMINI_API_KEY=your_key_here
MODEL_NAME=gemini-2.0-flash
TEMPERATURE=0.7"""

with open('.env', 'w') as f:
    f.write(env_content)

print(".env file created!")
print(env_content)
```

```python
# Load from .env (demo)
try:
    from dotenv import load_dotenv
    load_dotenv()
    import os
    print("Loaded from .env:")
    print(f"  MODEL_NAME: {os.getenv('MODEL_NAME')}")
    print(f"  TEMPERATURE: {os.getenv('TEMPERATURE')}")
except ImportError:
    print("Install python-dotenv first: pip install python-dotenv")
```
