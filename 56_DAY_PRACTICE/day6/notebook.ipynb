## 🐍 Why Virtual Environments?

**venv = Your project's personal bubble 🎈**
- Isolates packages for each project
- Prevents version conflicts
- Keep projects clean and separate!

## 2. Install Packages

```python
# After activating venv and installing packages in TERMINAL,
# run this to verify packages are installed:

import subprocess
import sys

# Check if key packages are installed
packages_to_check = ['google-generativeai', 'requests', 'dotenv']
print("📦 Checking installed packages...")
print("-" * 40)

for pkg in packages_to_check:
    try:
        __import__(pkg.replace('-', '_').replace('dotenv', 'dotenv'))
        print(f"✅ {pkg} is installed")
    except ImportError:
        print(f"❌ {pkg} NOT installed - run: pip install {pkg}")
```

## 4. Demo: Check Installed Packages

```python
# Demo: Check installed packages (run after pip install)
import subprocess

result = subprocess.run(['pip', 'list'], capture_output=True, text=True)
print("📦 All Installed Packages:")
print("-" * 40)
# Show first 20 packages
lines = result.stdout.split('\n')[:20]
for line in lines:
    print(line)
print("...")
print(f"\nTotal packages: {len(result.stdout.split())}")
```

## 5. Demo: python-dotenv

```python
# Create a .env file for API keys
env_content = """# API Keys - KEEP SECRET!
# Never commit .env to GitHub!
GEMINI_API_KEY=your_gemini_api_key_here
MODEL_NAME=gemini-2.0-flash
TEMPERATURE=0.7
MAX_TOKENS=2048
"""

# Write to file
with open('.env', 'w') as f:
    f.write(env_content)

print("✅ .env file created!")
print("\n📝 Contents:")
print(env_content)
print("\n⚠️  IMPORTANT: Add '.env' to your .gitignore!")
```

```python
# Load from .env file
try:
    from dotenv import load_dotenv
    load_dotenv()
    import os
    
    print("✅ Loaded from .env file:")
    print(f"   MODEL_NAME: {os.getenv('MODEL_NAME')}")
    print(f"   TEMPERATURE: {os.getenv('TEMPERATURE')}")
    print(f"   MAX_TOKENS: {os.getenv('MAX_TOKENS')}")
    print(f"   GEMINI_API_KEY: {'✅ Set' if os.getenv('GEMINI_API_KEY') else '❌ Not set'}")
    
except ImportError:
    print("❌ python-dotenv not installed!")
    print("   Run: pip install python-dotenv")
```
