# 🎬 Day 6: Classes - The Document Class

**Goal:** Build a Document class with all required methods!

---

## Your Document Class Code

```python
class Document:
    def __init__(self, title, content, metadata=None):
        self.title = title
        self.content = content
        self.metadata = metadata if metadata else {}
    
    def to_dict(self):
        return {"title": self.title, "content": self.content, "metadata": self.metadata}
    
    def word_count(self):
        return len(self.content.split())
    
    def char_count(self):
        return len(self.content)
    
    def summary(self, first_n=50):
        if len(self.content) <= first_n:
            return self.content
        return self.content[:first_n] + "..."
    
    @classmethod
    def from_json(cls, json_data):
        return cls(
            title=json_data.get("title", ""),
            content=json_data.get("content", ""),
            metadata=json_data.get("metadata", {})
        )
```

## Test Your Code

```python
# Test 1: Create a document
doc = Document("Test Doc", "Hello world, this is a test document!")
print("Title:", doc.title)
print("Content:", doc.content)
```

```python
# Test 2: word_count() and char_count()
print("Word count:", doc.word_count())
print("Char count:", doc.char_count())
```

```python
# Test 3: summary()
print("Summary (20):", doc.summary(20))
print("Summary (default):", doc.summary())
```

```python
# Test 4: to_dict()
print("As dict:", doc.to_dict())
```

```python
# Test 5: from_json() - class method
json_data = {"title": "JSON Doc", "content": "Created from JSON!", "metadata": {"author": "Krishna"}}
doc2 = Document.from_json(json_data)
print("From JSON - Title:", doc2.title)
print("From JSON - Content:", doc2.content)
print("From JSON - Metadata:", doc2.metadata)
```

- [ ] word_count() returns number of words
- [ ] char_count() returns number of characters
- [ ] summary(first_n) truncates text
