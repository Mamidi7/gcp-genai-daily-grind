## 1. Reading a Text File

```python
# Create a sample file first
with open("sample.txt", "w") as f:
    f.write("Hello, this is line 1\n")
    f.write("This is line 2\n")
    f.write("And this is line 3")

# Read the file
with open("sample.txt", "r") as f:
    content = f.read()

print("=== File Content ===")
print(content)
```

## 2. Reading Line by Line

```python
# Read line by line - useful for large files
with open("sample.txt", "r") as f:
    for line in f:
        print(f"Line: {line.strip()}")
```

```python
with open("whatthehlehd.txt","r") as f:
    for line in f:
        print(f"line: {line.split()}")
```


## 3. Writing to JSON

```python
import json

# Create a dictionary (like a JSON object)
config = {
    "model": "gemini-2.0-flash",
    "temperature": 0.7,
    "max_tokens": 1000,
    "top_p": 0.9
}

# Write to JSON file
with open("config.json", "w") as f:
    json.dump(config, f, indent=2)  # indent=2 makes it pretty

print("✅ config.json written!")
```


## 4. Reading from JSON

```python
# Read JSON file
with open("config.json", "r") as f:
    loaded_config = json.load(f)

print("=== Loaded Config ===")
print(f"Model: {loaded_config['model']}")
print(f"Temperature: {loaded_config['temp']}")
print(f"All: {loaded_config}")
print(loaded_config)
```

```python
with open("config.json","r") as f:
    config_loaded = json.load(f)

print("config loaded...")
print(f"Model: {config_loaded['model']}")
```

## 5. os module - File Operations

```python
import os

# Check if file exists
print("config.json exists?", os.path.exists("config.json"))

# Get current directory
print("Current directory:", os.getcwd())

# List files
print("Files in folder:", os.listdir("."))
```

## 6. Real-World Example: Save RAG Results

```python
# Simulate saving RAG evaluation results
rag_results = {
    "question": "What is RAG?",
    "answer": "RAG stands for Retrieval-Augmented Generation...",
    "sources": ["doc1.pdf", "doc2.pdf"],
    "faithfulness": 0.85,
    "answer_relevance": 0.92,
    "timestamp": "2026-03-03T10:30:00"
}

# Save to JSON
with open("rag_results.json", "w") as f:
    json.dump(rag_results, f, indent=2)

print("✅ RAG results saved to rag_results.json")
```
