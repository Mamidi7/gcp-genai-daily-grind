```python
projects = [
    {"name": "Data Pipeline", "days": 5},    # Row 1
    {"name": "ML Model", "days": 12},       # Row 2
    {"name": "API Deploy", "days": 3},       # Row 3
    {"name": "Dashboard", "days": 7}        # Row 4
]

long_projects = []
short_projects = []
for p in projects:
    if p["days"]>5:
        long_projects.append(p["name"])
    if p["days"]<=5:
        short_projects.append(p["name"])    
print(long_projects)
print(short_projects)

```
