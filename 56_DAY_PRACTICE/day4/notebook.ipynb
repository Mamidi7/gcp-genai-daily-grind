## Class = Movie Script, Object = Actor

```python
# Simple class example
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        return f"Hey, I'm {self.name}!"

p = Person("Krishna", 25)
print(p.name)
print(p.greet())
```

## 🛠️ Build the Document Class

```python
class Document:
    def __init__(self, title, content, metadata=None):
        self.title = title
        self.content = content
        self.metadata = metadata if metadata else {}
    
    def to_dict(self):
        return {
            "title": self.title,
            "content": self.content,
            "metadata": self.metadata
        }
    
    def word_count(self):
        return len(self.content.split())
    
    def has_keyword(self, keyword):
        return keyword.lower() in self.content.lower()

doc = Document(
    title="RAG Notes",
    content="A RAG system retrieves documents and generates answers using LLM.",
    metadata={"source": "lecture", "pages": 5}
)

print("Title:", doc.title)
print("Word count:", doc.word_count())
print("Has 'LLM'?", doc.has_keyword("LLM"))
print("Has 'AI'?", doc.has_keyword("AI"))
print("Dict:", doc.to_dict())
```

## 📝 Exercise: Add more methods

```python
# Add: char_count(), summary(first_n), from_json()

class Document:
    def __init__(self, title, content, metadata=None):
        self.title = title
        self.content = content
        self.metadata = metadata if metadata else {}
    
    def to_dict(self):
        return {"title": self.title, "content": self.content, "metadata": self.metadata}
    
    def word_count(self):
        return len(self.content.split())
    
    def char_count(self):
        return len(self.content)
    
    def summary(self, first_n=50):
        if len(self.content) <= first_n:
            return self.content
        return self.content[:first_n] + "..."
    
    @classmethod
    def from_json(cls, json_data):
        return cls(
            title=json_data.get("title", ""),
            content=json_data.get("content", ""),
            metadata=json_data.get("metadata", {})
        )

# Test
doc = Document("Test", "Hello world, this is a test document!")
print("Word:", doc.word_count(), "Char:", doc.char_count())
print("Summary:", doc.summary(20))

json_data = {"title": "JSON Doc", "content": "From JSON!"}
doc2 = Document.from_json(json_data)
print("From JSON:", doc2.title, doc2.content)
```
