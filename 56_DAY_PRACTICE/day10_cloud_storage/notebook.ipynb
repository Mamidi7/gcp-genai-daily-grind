## Part 4: BigQuery Window Functions (SQL Track)

```python
# BigQuery Window Functions

# ROW_NUMBER - Assigns unique sequential numbers
row_number_example = """
-- Rank employees by salary within each department
SELECT 
    name,
    department,
    salary,
    ROW_NUMBER() OVER (PARTITION BY department ORDER BY salary DESC) as salary_rank
FROM employees
"""

# RANK - Handles ties (same rank for tied values)
rank_example = """
-- Rank products by sales (ties share rank)
SELECT 
    product_name,
    sales,
    RANK() OVER (ORDER BY sales DESC) as sales_rank
FROM products
"""

# LAG/LEAD - Previous/Next row values
lag_lead_example = """
-- Compare each month's revenue with previous month
SELECT 
    month,
    revenue,
    LAG(revenue) OVER (ORDER BY month) as prev_month_revenue,
    revenue - LAG(revenue) OVER (ORDER BY month) as growth
FROM monthly_sales
"""

print("=== BigQuery Window Functions ===\n")
print("1. ROW_NUMBER (unique ranking):")
print(row_number_example)
print("\n2. RANK (handles ties):
")
print(rank_example)
print("\n3. LAG/LEAD (previous/next row):")
print(lag_lead_example)
```

interview_qa = """
Q1: What does PARTITION BY user_id do?
A: Splits data into groups by user. Each user gets their own ranking.

Q2: Why use ROW_NUMBER() instead of RANK()?
A: ROW_NUMBER = unique ranks (1,2,3,4)
   RANK = same rank for ties (1,1,3)
   For "one row per user", use ROW_NUMBER.

Q3: When to use DENSE_RANK?
A: When you want consecutive ranks without gaps.
   RANK: 1,1,3  vs  DENSE_RANK: 1,1,2

Q4: LAG vs LEAD?
A: LAG = previous row
   LEAD = next row

Q5: Why useful for AI/ML backend?
A: Track user patterns, detect latency issues, find retry patterns.
"""

-- Q2: Rank pages by views (using RANK)

  page.pagePath,
  totals.pageviews,

FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`,
  UNNEST(hits) AS page
ORDER BY page_rank;

-- Q3: Session sequence (using LAG)

  visitId,
  visitStartTime,
  LAG(visitStartTime) OVER (ORDER BY visitStartTime) as prev_session
FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
LIMIT 10;
"""
