```python

from pathlib import Path
import json

config_path = Path("./config/runtime_profile.json")

if not config_path.exists():
    print("File missing:", config_path)
    data = {}
else:
    with config_path.open("r", encoding="utf-8") as f:
        data = json.load(f)

if not isinstance(data, dict):
    raise ValueError("Runtime config must be a JSON object")

print("Loaded OK")
print("Type:", type(data).__name__)
print("Keys:", sorted(data.keys()))
print("runtime_profile:", data.get("runtime_profile", "missing"))
```

### Goal

### You Must Cover
- Mask full credentials path
- Never print raw key/token
- Return only safe fields

```python

from pathlib import Path

def mask_path(v: str) -> str:
    if not v:
        return "not-set"
    return f".../{Path(v).name}"

print(mask_path("/Users/me/secrets/my-sa-key.json"))
print(mask_path(""))
```

## Topic 4 - Missing Env Fail-Fast

### Goal
Detect required env issues before request-time failures.

### You Must Cover
- Required key check for `GCP_PROJECT_ID`
- Clear error message
- Status mapping (`invalid` / `valid`)

```python
import os

def validate_required_env() -> list[str]:
    errors = []
    if "GCP_PROJECT_ID" not in os.environ:
        errors.append("GCP_PROJECT_ID is missing from environment")
    return errors

print(validate_required_env())
```
