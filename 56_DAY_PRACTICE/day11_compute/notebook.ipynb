## Part 3: BigQuery - Subqueries & CTEs (SQL Track)

```python
# Subqueries vs CTEs

# Subquery - Nested query
subquery_example = """
-- Find products that cost more than average
SELECT product_name, price
FROM products
WHERE price > (SELECT AVG(price) FROM products)
"""

# CTE - More readable, reusable
cte_example = """
-- Same query using CTE
WITH avg_price AS (
    SELECT AVG(price) as avg FROM products
)
SELECT product_name, price
FROM products, avg_price
WHERE price > avg_price.avg
"""

# Correlated Subquery
correlated_example = """
-- Get each department's highest earner
SELECT name, salary, department
FROM employees e1
WHERE salary = (
    SELECT MAX(salary) 
    FROM employees e2 
    WHERE e2.department = e1.department
)
"""

print("=== Subqueries vs CTEs ===\n")
print("1. Subquery (nested):")
print(subquery_example)
print("\n2. CTE (Common Table Expression):")
print(cte_example)
print("\n3. Correlated Subquery:")
print(correlated_example)

print("\n=== When to use what? ===")
print("CTE: Complex queries, better readability, recursive")
print("Subquery: Simple filters, in WHERE clause")
print("Correlated: Row-by-row comparisons")
```
